**Splitting X and Y values**

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

**Train-Test-Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

**Model Building:**



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.preprocessing import StandardScaler # Import for standard scaling of the data
from sklearn.preprocessing import MinMaxScaler # Import for standard scaling of the data
```

```python
# standard scale data
ss = StandardScaler()
X_train = ss.fit_transform(x_train)
X_test = ss.transform(x_test)
```

```python
# tested MinMaxScaler as KNN historically does better with MinMax
mm = MinMaxScaler()
X_train_mm = ss.fit_transform(x_train)
X_test_mm = ss.transform(x_test)
```

**KNN**

```python
#Train
knn = KNeighborsClassifier()
knn.fit(X_train_mm, y_train)
```

```python
y_pred_knn = knn.predict(X_test_mm)
```

```python
score = accuracy_score(y_test, y_pred_knn)
print('Accuracy Score:',score*100)
report = classification_report(y_test, y_pred_knn)
print('Classification Report:', report)
```

```python
#Test
knn1 = KNeighborsClassifier()
knn1.fit(X_test, y_test)
```

```python
y_pred_knn1 = knn1.predict(X_test)
```

```python
score1 = accuracy_score(y_test, y_pred_knn1)
print('Accuracy Score:',score1*100)
report1 = classification_report(y_test, y_pred_knn1)
print('Classification Report:', report1)
```

**Decision Tree**

```python
#Train
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred_dtc = dtc.predict(X_test)
```

```python
score_dtc = accuracy_score(y_test, y_pred_dtc)
print('Accuracy Score:',score_dtc*100)
report_dtc = classification_report(y_test, y_pred_dtc)
print('Classification Report:', report_dtc)
```

**SVM**

```python
#Train
svm = SVC()
svm.fit(X_train, y_train)
```

```python
y_pred_svm = svm.predict(X_test)
```

```python
score_svm = accuracy_score(y_test, y_pred_svm)
print('Accuracy Score:',score_svm*100)
report_svm = classification_report(y_test, y_pred_svm)
print('Classification Report:', report_svm)
```

```python
#test
svm1 = SVC()
svm1.fit(X_test, y_test)
```

```python
y_pred_svm1 = svm1.predict(X_test)
```

```python
score_svm1 = accuracy_score(y_test, y_pred_svm1)
print('Accuracy Score:',score_svm1*100)
report_svm1 = classification_report(y_test, y_pred_svm1)
print('Classification Report:', report_svm1)
```

```python
class_name = ("KNN-Train","KNN-Test","Decision Tree","SVM-Train","SVM-Test")
class_score = (score,score1,score_dtc,score_svm,score_svm1)

y_pos = np.arange(len(class_score))
colors = ("red","gray","purple","green","orange")

plt.figure(figsize=(10,5))

barwidth = 0.2
barposition = y_pos - barwidth / 2
bars = plt.bar(y_pos, class_score, color= colors, width=barwidth)

for bar, score in zip(bars, class_score):
  plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.01,f"{score:.2f}", fontsize=12)

plt.xticks(y_pos, class_name, fontsize=9)
plt.ylabel('Accuracy')
plt.grid()
plt.title("Accuracy Score Comparsion of the Classed", fontsize=10)
plt.show()
```
