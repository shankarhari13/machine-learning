**LabelEncoding for categorical data**

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df1['Sex'] = lb.fit_transform(df1['Sex'])
df1['ChestPainType'] = lb.fit_transform(df1['ChestPainType'])
df1['RestingECG'] = lb.fit_transform(df1['Sex'])
df1['ExerciseAngina'] = lb.fit_transform(df1['ExerciseAngina'])
df1['ST_Slope'] = lb.fit_transform(df1['ST_Slope'])
```

```python
df1.head(3)
```

**Train_Test_Split**

```python
# Create X and Y Variables
X = df1.iloc[:, :-1]
Y = df1.iloc[:, -1]
```

**Standardization**

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit_transform(X)
```

```python
# Split the X and Y dataset into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 11)
```

**Model Building**

**1. Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train,Y_train)
```

```python
#Predicting x_test data
y_pred = log.predict(X_test)
```

**Evaluation Metrics**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm=confusion_matrix(Y_test,y_pred)
score=accuracy_score(Y_test,y_pred)*100
print('Error value:',np.mean(y_pred!=Y_test))
print('Confusion Matrix:',cm)
print('Accuracy Score:', score)
print('Classification Report:', classification_report(Y_test,y_pred))
```

**Confusion Matrix**

```python
sns.heatmap(cm,cbar = True,  square = True, annot=True, fmt= '.2f', annot_kws={'size': 12}, cmap= 'coolwarm')
```

**2. Decision Tree Classification**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=11)
```

```python
dtc.fit(X_train,Y_train)
```

```python
y_pred1 = dtc.predict(X_test)
```

**Evaluation Metrics**

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm1=confusion_matrix(Y_test,y_pred1)
score1=accuracy_score(Y_test,y_pred1)*100
print('Error value:',np.mean(y_pred1!=Y_test))
print('Confusion Matrix:',cm1)
print('Accuracy Score:', score1)
print('Classification Report:', classification_report(Y_test,y_pred1))
```

**Confusion Matrix**

```python
sns.heatmap(cm1,cbar = True,  square = True, annot=True, fmt= '.2f', annot_kws={'size': 12}, cmap= 'coolwarm')
```

**Comparsion of all Accuracy Score**

```python
class_name = ("Logistic Regression","Decision Tree","Naive Bayes","SVM")
class_score = (score,score1,score2,score3)

y_pos = np.arange(len(class_score))
colors = ("gray","purple","green","orange")

plt.figure(figsize=(10,5))

barwidth = 0.2
barposition = y_pos - barwidth / 2
bars = plt.bar(y_pos, class_score, color= colors, width=barwidth)

for bar, score in zip(bars, class_score):
  plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 0.01,f"{score:.2f}", fontsize=12)

plt.xticks(y_pos, class_name, fontsize=9)
plt.ylabel('Accuracy')
plt.grid()
plt.title("Accuracy Score Comparsion of the Classed", fontsize=10)
plt.show()
```
